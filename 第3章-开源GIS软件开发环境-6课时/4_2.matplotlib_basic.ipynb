### matplotlib绘图基础
介绍画布figure, 子图axes, 子图坐标系Axis，子图标题title等。


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### 序列数据可视化
data = [1,2,3,4]
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].plot(data, marker='.', color='red', markersize=10)
```

```python
### 散点数据可视化
data_x = [1,2,3,4]
data_y = [3,2,1,4]
fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.scatter(data_x, data_y, s=60, marker='.', color='green')
```

```python
### 柱状图
fig, ax = plt.subplots(1,1, figsize=(5,4))
data_student = ['zhang1', 'zhang2', 'zhang3', 'zhang4']
data_scores = [87,92,79,82]
ax.bar(data_student, data_scores, width=0.4, color='red')
```

```python
### 图像数据可视化
# arr = np.random.randn(100,100)
arr = np.linspace(0, 1, 100)
arr_2d = np.expand_dims(arr, 1).repeat(100, axis=1)    ## 1.扩充维度；2.复制数据
arr_2d.shape
fig, ax = plt.subplots(1,1, figsize=(5,5))
ax.imshow(arr_2d, cmap='grey')
```
